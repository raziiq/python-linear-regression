# Regression Models in Python
## Simple Linear and Multiple Linear Regression

```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# import the dataset
df = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv")
df.head()
```

### Simple Linear Regression

```python
# Create an object of Linear Regression
lr = LinearRegression()
```

```python
# Lets prepare the X (Independent variable) and Y (Dependent variable)
X = df[['wheel-base']]
Y = df['price']
```

```python
lr.fit(X, Y)
```

```python
Yhat = lr.predict(X)
Yhat[0:5]
```

```python
# Lets find out the Y intercept
lr.intercept_
```

```python
# Lets see the slope variable
lr.coef_
```

```python
# import the visualization package: seaborn
import seaborn as sns
%matplotlib inline 
```

```python
# Lets find out the regression using regplot from seaborn
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.regplot(x="wheel-base", y="price", data=df)
plt.ylim(0,)
```

```python
# Lets see how this feature does with Residual plot
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.residplot(df['wheel-base'], df['price'])
plt.show()

# The residual plot does not show a random spread which is not a good sign for the model
```

```python
# Next step is to find the R-squared and Mean Square Error for this model.
# These values for each model will be compared to see which has model has the most appropriate values

lr.fit(X, Y)
# Find the R^2
print('The R-square is: ', lr.score(X, Y))
```

```python
# Lets find MSE. But first we need to find Yhat
Yhat=lr.predict(X)
print('The output of the first four predicted value is: ', Yhat[0:4])
```

```python
# Finding MSE
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(df['price'], Yhat)
print('The mean square error of price and predicted value is: ', mse)
```

#### So the R-squared is 0.3418060603419337 and MSE is  41361996.58923002 for Wheel-Base
